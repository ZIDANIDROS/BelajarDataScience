```python
import pandas as pd
import numpy as np

data = {
    "Nama Produk": [
        "Gamis Hijab", "Jilbab Syar'ih", "Cadar", "Gamis Pesta",
        "Jilbab Modern", "Cadar Premium", "Gamis Kasual",
        "Cadar Sederhana", "Jilbab Lengan Panjang", "Gamis Anak"
    ],
    "Jumlah Penjualan": [50, 30, 20, 40, 10, 25, 35, 15, 45, 28], # Benefit
    "Persentase Penurunan Stok (%)": [15, 7, 2, 20, 13, 10, 5, 8, 18, 3], # Cost
    "Waktu Stok (hari)": [10, 30, 20, 5, 15, 12, 18, 25, 7, 14], # Cost
    "Margin Keuntungan (%)": [20, 25, 15, 18, 30, 22, 15, 12, 28, 19], # Benefit
    "Ketersediaan Restock": [8, 5, 10, 7, 6, 9, 8, 4, 7, 6] # Benefit
}


df = pd.DataFrame(data)
df
```

```python
# Mengubah nama produk menjadi A1-A10
df.index = [f"A{i+1}" for i in range(len(df))]

# Mengubah nama kolom menjadi C1-C5
df.columns = ["Nama Produk", "C1", "C2", "C3", "C4", "C5"]

# Menampilkan hasil
df
```

```python
weights = np.array([0.3, 0.25, 0.2, 0.15, 0.1])  # Bobot untuk Jumlah Penjualan, Persentase Penurunan Stok, Waktu Stok, Margin Keuntungan
weights
```

```python
# Matriks normalisasi R'
R = df[["C1", "C2", "C3", "C4", "C5"]] / np.sqrt((df[["C1", "C2", "C3", "C4", "C5"]] ** 2).sum())

R
```

```python
# Bobot untuk setiap kriteria
weights = np.array([0.3, 0.25, 0.2, 0.15, 0.1])
```

```python
# Mengalikan matriks normalisasi dengan bobot Y
Y = R * weights

# Menampilkan hasil
print("Matriks Normalisasi Terbobot (C1-C5):")
Y
```

```python
# Menghitung nilai ideal positif (A+)
A_plus = [
    Y.max()[0],  # max for C1 (Benefit)
    Y.min()[1],  # min for C2 (Cost)
    Y.min()[2],  # min for C3 (Cost)
    Y.max()[3],  # max for C4 (Benefit)
    Y.max()[4]   # max for C5 (Benefit)
]

# Menghitung nilai ideal negatif (A-)
A_minus = [
    Y.min()[0],  # min for C1 (Benefit)
    Y.max()[1],  # max for C2 (Cost)
    Y.max()[2],  # max for C3 (Cost)
    Y.min()[3],  # min for C4 (Benefit)
    Y.min()[4]   # min for C5 (Benefit)
]
```

```python
print("Ideal Positif (A+):")
A_plus
```

```python
print("Ideal Negatif (A-):")
A_minus
```

```python
# Menghitung D+ dan D-
D_plus = np.sqrt(((Y - A_plus) ** 2).sum(axis=1))
D_minus = np.sqrt(((Y - A_minus) ** 2).sum(axis=1))

# Menampilkan hasil
print("D+:\n", D_plus)
print("===================")
print("D-:\n", D_minus)
```

```python
# Menghitung nilai preferensi (P)
P = D_minus / (D_plus + D_minus)

# Menambahkan nilai preferensi ke DataFrame
df['Preferensi'] = P

# Menampilkan hasil
df
```

```python
# Membuat ranking
df['Rank'] = df['Preferensi'].rank(ascending=True)
df_sorted = df.sort_values(by='Rank', ascending=True)

# Menampilkan hasil
df_sorted[['Nama Produk', 'Preferensi', 'Rank']]
```
